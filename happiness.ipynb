# Preguntas

## Esta relacionado las ingresos per capita con la felicidad?
*Aqui exponemos la primera duda sobre si el ingresos per capita de los pasises esta relacionado con la felicidad hay una tendencia no tan clara pero esto me genera mas dudas*
*Los paises mas libres son mas felices*

```python
sns.relplot(data=happiness21, x='Ladder score',y='Logged GDP per capita',hue='Regional indicator');
```

## La libertad esta relacionada con la felicidad
*La hipotesis de la felicidad van relcionada no muestra una relacion positiva perfecta ya que algunos paises esta perfectamente en la tendencia pero tenemos algunos outlayers del sur de asia, sub-sahara y el sur de asia*

Esto nos muestra que la felicidad "no simpre" esta relaciona al ingreso per capita y a la libertdad por los dato abordados por la muestra

```python
sns.relplot(data=happiness21, x="Freedom to make life choices",y='Ladder score',hue='Regional indicator');
```

## Los paisess mas ricos son los mas felices 
La hipotesis era que los paisess mas ricos "eran  100% los mas felices" pero el analisis nos muestras que este no es tan necesarion hay pasises que son muy ricos y no tan felices como otrs que son menos mas ricos pero mas felices. La conclusion que un pais rico no indica al 100% que se mas felices que otros un poco mas probres 

```python
slices=happiness21.loc[:,['Country name','Logged GDP per capita','Ladder score','Regional indicator']]
slices.sort_values(by=['Logged GDP per capita','Country name','Ladder score','Regional indicator'],ascending=False)
sns.relplot(data=slices, x='Logged GDP per capita',y='Ladder score',hue='Regional indicator',height=8,size='Logged GDP per capita');
```
